### Data exploration

```python
import pandas as pd
from sklearn.datasets import load_iris
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "notebook"
```

## Load dataset

```python
iris = load_iris(as_frame=True)
df = iris.frame

print(df.head(5))
```

```python
print(df['target'].unique())
```

Plot a histogram for each column

```python
for column in df.columns[:-1]:

    fig = go.Figure()

    fig.add_trace(
        go.Histogram(x=df[column]))

    fig.update_layout(
        title=column,
    )

    fig.show()
```

Plot a scatterplot for each pair of columns

```python
columns = df.columns[:-1]
for i, x_col in enumerate(columns):
    print("x_column", x_col, i)
    for j, y_col in enumerate(columns[i+1:]):
        print("y_col:", y_col, j+i+1)


        fig=go.Figure()
        fig.add_trace(
            go.Scatter(x=df[x_col], y=df[y_col],  mode="markers",))
        fig.update_layout(
    xaxis_title=x_col,
    yaxis_title=y_col,
    title=f"{x_col} vs {y_col}"
)
        fig.show()
```
